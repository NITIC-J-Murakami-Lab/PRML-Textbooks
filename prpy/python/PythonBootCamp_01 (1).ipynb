# 「fizzbuzzを行う関数を実装しよう」



fizzbuzzを行う関数を実装しましょう。
Fizz Buzzについては[wikiを参照](https://ja.wikipedia.org/wiki/Fizz_Buzz)してください。


関数の雛形として、以下を用意しました。
```python
def fizzbuzz(max_number):
    for i in range(max_number):
        print("{}".format(i))
```
これを拡張して、max_numberとして与えられた数字(int)までfizz buzzする関数を作成し、実行結果を示してください。


````python
def fizzbuzz(num):
    for num in range(1, 101):
        if num % 3 == 0 and num % 5 == 0:
            print('fizzbuzz')
        elif num % 3 == 0:
            print('fizz')
        elif num % 5 == 0:
            print('buzz')
        else:
            print(num)

fizzbuzz(1)
````

# 「フィボナッチ数列を100桁目まで返す関数を実装しよう」


フィボナッチ数列を列挙する関数を作りましょう。

```python
fib_list = [1,1]
```
が最初に与えられます。これを元にしてどんどんfib_listにフォボナッチ数列を作成して、len(fib_list)==100になるまでそれを繰り返してください。

````python
def fibo(n):
    if n < 3:
        return [1,1]
    fibo_list = [1,1]
    for i in range(2, n-1):
        fibo_i = fibo_list[i-1] + fibo_list[i-2]
        fibo_list.append(fibo_i)
    return fibo_list
        
print(fibo(10))
````

````python
fib_list = [1,1]
def my_fib(n):
  if n <= 2:
    raise ValueError
    #return [1,1]
  for i in range(2, n-1):
    fib_list.append(fib_list[-1] + fib_list[-2])
  return fib_list
````

````python
my_fib(1)
````

# 「aからbまでの和を計算する関数を実装しよう」

a=5, b =10ならば
$$
result = 5+6+7+8+9+10
$$

のように、aとbの２つのint型数字を受け取って、aからbまで足し合わせた答えを返す関数を作りましょう。

ただし、以下の条件をみたすようにしてください。
* a < bの場合は正しく動作し、それ以外の場合（a >= b）は標準出力に"ERROR"と表示。
* Pythonの組み込み関数であるsum()は使わない事。


````python
def sum(a,b):
    total = 0
    if a >= b:
        print("ERROR")
    else:
        for num in range(a,b+1):
            total += num
        return total

sum(5,10)
````

````python
sum(10,5)
````

# 「階乗を計算する関数を実装しよう」



 [階乗とは](https://ja.wikipedia.org/wiki/%E9%9A%8E%E4%B9%97)

階乗を実装する関数を作成しましょう。

```python
def factorial(target_number):
    result = 1
    return result
```
上記の関数を元にして、階乗を計算する関数を完成させましょう。

````python
def fact(n):
    if n == 0:
        return 1
    else:
        return n * fact(n-1)

fact(5)
````

````python
target_number = 5
y = range(1, target_number+1)
y = reversed(y)

result = 1
for item in y:
  result *= item

result
````

ユークリッドの互除法を使って、２つの数字から最大公約数を求める関数を作成しましょう！

ただし、これは非常に簡単なコードになります。ですので、ユークリッドの互除法について知りたい場合は[このページ](https://ja.wikipedia.org/wiki/%E3%83%A6%E3%83%BC%E3%82%AF%E3%83%AA%E3%83%83%E3%83%89%E3%81%AE%E4%BA%92%E9%99%A4%E6%B3%95)以外のみから理解して、プログラムを実装してください。



````python
def euclidean_algorithm(a, b):
    while a % b != 0:
        c = a % b
        #a = b
        #b = c
        a,b = b,c
    return b
    
euclidean_algorithm(28, 20)
````

````python
def t(a, b):
    for i in range(10000):
        if a % b == 0:
            c = a % b
            a = b
            b = c
    return b

t(28, 20)
````
