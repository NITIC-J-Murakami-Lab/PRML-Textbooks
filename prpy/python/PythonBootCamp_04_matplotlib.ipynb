# 「乱数でサンプルデータを作ろう」

ランダムに0～1の範囲から取り出された乱数を20個まとめたリストを作りましょう。

ここで「ランダムに0～1の範囲から取り出された乱数」とは__「 $0 \leq x \leq 1$の範囲の一様乱数からサンプリングされた数字」__を指します。

randomモジュールを使って一様乱数を使う方法を以下に示します。

```python
import random

random.uniform(0,1)
```

※random.uniform(最小値,最大値)です。

では課題を解くために、以下の関数を完成させてください。

完成したらこれを実行してsample_list変数に返り値を入れてください。

```python
def sample_list_generator(min_val, max_val, sample_size)->list:
    """
    min_valは乱数の最小値
    max_valは乱数の最大値
    sample_sizeはサンプルリストのサイズ
    """
    sample_list = [] # とりあえずからのリストを作成
    for i in range(sample_size):
      x = random.uniform(min_val, max_val) # 第一引数～第２引数の間からランダムに数字を取り出す関数
      sample_list.append(x) # appendはlistに要素を追加するメソッド
    return sample_list # :list
```

```python
sample_list_generator(0,1,10)
```

```python
iris = load_iris()
iris.keys()
```

```python
#x = [x[0] for x in iris.data]
#y = [x[1] for x in iris.data]

x = iris.data[:,0] # : = 全部
y = iris.data[:,1] #
plt.scatter(x, y, c=iris.target)
```

```python
iris.target # 各データのクラス
```

### 問題1 
pltを使って、散布図を作ってみよう。データは0~3までの特徴を持っている。そのうちの2と3の特徴を使おう。クラスごとに色もつけよう。



```python
from sklearn.decomposition import TruncatedSVD as LSI

encoder_model = LSI(n_components=2)
embed = encoder_model.fit_transform(iris.data)

embed.shape
```

```python
plt.scatter(embed[:,0], embed[:,1], c=iris.target)
```
