

例えばスパムメールフィルタリングを例に考えてみましょう．
- $P(A)$... 去年一年間で計算した「あるメールがスパムメールである確率」
- $P(B)$... 全てのメールの中で「当選」という言葉が入っているメールの割合
- $P(A|B)$... 「当選」という言葉が入っているメールの中で，迷惑メールである確率
- $P(B|A)$...　スパムメールの中で，「当選」という言葉が入っているメールの確率

この時，『$P(B|A)$...　スパムメールの中で，「当選」という言葉が入っているメールの確率』は以下のように表すことができ，これをベイズの定理と呼ぶ．  
$$
P(B|A) = \frac{P(B)\cdot P(A|B)}{P(A)}
$$


## 自己情報量 (Entropy)

$$
i(x)=−\log_{2}{P(x)}
$$

::::{note}  
$i(x)=−\log_{n}{P(x)}$おいて対数の底$n$を
- ２にとれば情報量の単位はビット
- 10にとればデジット
- ネイピア数$e$ にとればナットになります．  

ナット：nat, NAtural unit of Information"（情報の自然単位）
> ボルツマン定数を1に正規化した自然単位系では、熱力学エントロピーの値はナットで測定できる。情報量を自然対数で書き表すと、
> $$
i(x)=−\ln{P(x)}$$  
> となり、ナットによる数値を暗黙のうちに与えている。  
引用：[Wikipedia](https://ja.wikipedia.org/wiki/%E3%83%8A%E3%83%83%E3%83%88_(%E5%8D%98%E4%BD%8D))

::::

エントロピーは確率が1のとき0となり，確率が0のとき無限大に発散する．

```python
proba = np.linspace(0,1, 1000)

entropy = lambda p: -np.log(p)

def build_plot1():
    fig = plt.figure()
    ax = fig.add_subplot()
    ax.plot(proba, entropy(proba))
    fig.suptitle("EntropyとProbabilityの関係")
    ax.set_xlabel("probability")
    ax.set_ylabel("entropy")

    # 補助目盛を追加
    ax.minorticks_on()

    # 主目盛線と補助目盛線を表示
    ax.grid(which="both")
    plt.close()
    return fig

build_plot1()
```
