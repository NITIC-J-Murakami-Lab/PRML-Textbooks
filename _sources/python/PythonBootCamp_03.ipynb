# 「面の数を変えられるサイコロを実装した関数を作ろう」

先程は六面に限定していましたが、今度は面の数を自由に変えられるようにしましょう。  

目の数はn_rollで決定します。  
それ以外は先程の関数と同様にしましょう。  

````python
def mutable_dice(n_roll:int)->int:
    num = random.randint(1, n_roll)
    return num
````
